```python
df = pd.read_csv('Downloads/animacy.csv', sep=',')
print(df.columns.tolist())
```

```python
features = ['manmade_mean', 'precious_mean', 'lives_mean', 'heavy_mean',
            'natural_mean', 'moves_mean', 'grasp_mean', 'hold_mean',
            'be.moved_mean', 'pleasant_mean', 'arousing_mean']

df_features = df[features]

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_features.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation between semantic features')
plt.tight_layout()
plt.show()
```
